```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

MutexData=np.array([10.28, 5.47, 3.72, 3.10, 2.62, 2.39, 2.47, 2.21, 2.22, 2.17, 2.05, 2.51, 2.28, 2.37])
AtomicData=np.array([10.07, 5.18, 3.60, 2.70, 2.24, 1.97, 1.77, 2.63, 1.76, 2.78,15.33, 1.95, 7.89, 6.79])
SequentialData=np.array([10.18,10.18,10.18,10.18,10.18,10.18,10.18,10.18,10.18,10.18,10.18,10.18,10.18,10.18])

SpeedupM=10.18/MutexData
SpeedupA=10.18/AtomicData
SpeedupS=10.18/SequentialData

  # data
mdf = pd.DataFrame({
  'x_axis': range(1,15),
  'y_axis': SpeedupM
})

adf = pd.DataFrame({
  'x_axis': range(1,15),
  'y_axis': SpeedupA
})

sdf = pd.DataFrame({
  'x_axis': range(1,15),
  'y_axis': SpeedupS
})

  # plot
plt.plot('x_axis', 'y_axis', data=mdf, linestyle='-', marker='o', label='mutex')
plt.plot('x_axis', 'y_axis', data=adf, linestyle='-', marker='o', label='atomic')
plt.plot('x_axis', 'y_axis', data=sdf, linestyle='-', marker='o', label='sequential')
plt.legend(loc="upper left")
plt.xlabel("number of processors")
plt.ylabel("processing time")
plt.show()
```
